# Bio Safe vs Misuse Dataset Peek

Quick inspection of the first few safe (label=1) and harmful (label=0) entries from `artifacts/bio_safe_misuse.jsonl`.


```python
from pathlib import Path
import json
import pandas as pd

DATASET_PATH = Path("../artifacts/bio_safe_misuse.jsonl")

if not DATASET_PATH.exists():
    raise FileNotFoundError(f"Dataset not found at {DATASET_PATH}")

records = []
with DATASET_PATH.open("r", encoding="utf-8") as f:
    for line in f:
        obj = json.loads(line)
        records.append(obj)

df = pd.DataFrame(records)
label_counts = df["label"].value_counts().sort_index()
label_counts
```

```python
harmful = df[df["label"] == 0].head(5)["statement"].reset_index(drop=True)
safe = df[df["label"] == 1].head(5)["statement"].reset_index(drop=True)

harmful_df = pd.DataFrame({"Harmful (label=0)": harmful})
safe_df = pd.DataFrame({"Safe (label=1)": safe})

harmful_df, safe_df
```
